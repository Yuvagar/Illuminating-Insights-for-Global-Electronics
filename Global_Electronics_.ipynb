```python
# checking any null values
cust_data.isnull().sum()
```

```python
# Data convesion of DOB
cust_data['Birthday']=pd.to_datetime(cust_data['Birthday'],format='%m/%d/%Y').dt.date
```

```python
# Calculate and create table of age by using date time
today = datetime.today()
cust_data["Age"] = cust_data["Birthday"].apply(
    lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day))
)
```

```python
# Fill null State code
cust_data['State Code']=cust_data['State Code'].fillna('NP')
```
